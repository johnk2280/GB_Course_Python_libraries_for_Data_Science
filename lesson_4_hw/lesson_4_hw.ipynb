В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.Дайте название графикам: 'График y1', 'График y2' и т.д.

#### Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

fig.set_size_inches(8, 6)

ax[0][0].plot(x, y1)
ax[0][0].axis([-5, 5, 0, 9])

ax[0][1].plot(x, y2)
ax[1][0].plot(x, y3)
ax[1][1].plot(x, y4)
```

#### Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

fig.set_size_inches(8, 6)
fig.subplots_adjust(hspace=0.3, wspace=0.3)

ax[0][0].plot(x, y1)
ax[0][0].axis([-5, 5, 0, 9])

ax[0][1].plot(x, y2)
ax[1][0].plot(x, y3)
ax[1][1].plot(x, y4)
```

С помощью функции numpy.where, используя матрицу D_inv в качестве маски, а матрицы B и C - в качестве источников данных, получите матрицу E размером 4x4.  Элементы матрицы E, для которых соответствующий элемент матрицы D_inv равен 1, должны быть равны соответствующему элементу матрицы B, а элементы матрицы E, для которых соответствующий элемент матрицы D_inv равен 0, должны быть равны соответствующему элементу матрицы C.
